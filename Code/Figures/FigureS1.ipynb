# Figure 2

In this notebook we will generate Figure 2. 
It shows the distribution of values for the covariates age, height, weight, and BMI, in the three different populations from fineStructure, divided by sex

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import os
```

We'll load the tot_samples datasets, containing only the set of 2558 individuals

```python
#Folders
projpath = os.path.realpath("../..")
pathres  = os.path.join(projpath, "Results")
pathfig  = os.path.join(projpath, "Results", "Figures")
```

```python
#Read Databases
os.chdir(pathres)
dat = pd.read_csv("tot_samples.csv")
```

Now we'll generate the figure

```python
#Defining some functions
def remove_labels(ax, title):
    ax.get_legend().remove()
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_title(title, fontsize = 10)

#Plot
fig = plt.figure(dpi = 250, figsize = [10, 6])
ax1  = fig.add_subplot(2,2,1)
ax2  = fig.add_subplot(2,2,2)
ax3  = fig.add_subplot(2,2,3)
ax4  = fig.add_subplot(2,2,4)
traits = ['Age', 'Height', 'Weight', 'BMI']
titles = ['Age', 'Height (cm)', 'Weight (kgs)', 'BMI']

i = 0
for ax in [ax1, ax2, ax3, ax4]:
    g = sns.violinplot(x = 'cluster_final', y = traits[i], hue = 'Sex', data = dat, palette = 'Set2', ax = ax)
    handles, labels = ax.get_legend_handles_labels()
    remove_labels(ax, titles[i])
    
    i = i + 1

fig.legend(handles, labels, loc='center right')
os.chdir(pathfig)
plt.savefig("FigureS1.png")
plt.show()
```
